```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

```python
#- 남성 흡연자 조건 필터링
#- 팁을 가장 많이 낸 순서로 내림차순 정렬
#- 상위 10개 출력
tips['smoker'] == 'Yes'
condition = tips['smoker'] == 'Yes'
tips[condition].head()
smoker = tips[tips['smoker'] == 'Yes']
cond1 = tips.sort_values(['tip'], ascending=[False]).head(10)
tips.loc[smoker | cond2].head(10)
```

```python
#식사인원이 5명 이상 저녁테이블,흡자가 없는 주말테잉블, 토탈빌, 팁 출력
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
```
